```python
import gmsh
import os
from pathlib import Path
```

# Function libs

```python
def generate_refined_tet_mesh(
    step_path: str,
    out_msh_path: str,
    global_size: float = 3.0,
    contact_size: float = 0.75,
    contact_box_min=(-10.0, -10.0, -10.0),
    contact_box_max=(10.0, 10.0, 10.0),
    msh_version: float = 4.1,
    show_gui: bool = False,
):
    gmsh.initialize()
    gmsh.option.setNumber("General.Terminal", 1)

    gmsh.model.add("finray_refined_mesh")

    step_path = str(Path(step_path).resolve())
    out_msh_path = str(Path(out_msh_path).resolve())

    # 1.Read STEP file and create geometry
    gmsh.model.occ.importShapes(step_path)
    gmsh.model.occ.synchronize()

    # 2. Set global mesh size
    gmsh.option.setNumber("Mesh.CharacteristicLengthMin", contact_size)
    gmsh.option.setNumber("Mesh.CharacteristicLengthMax", global_size)

    # 3. Construct a box field to refine mesh in the contact region
    field = gmsh.model.mesh.field.add("Box")

    gmsh.model.mesh.field.setNumber(field, "VIn", contact_size)
    gmsh.model.mesh.field.setNumber(field, "VOut", global_size)

    gmsh.model.mesh.field.setNumber(field, "XMin", contact_box_min[0])
    gmsh.model.mesh.field.setNumber(field, "YMin", contact_box_min[1])
    gmsh.model.mesh.field.setNumber(field, "ZMin", contact_box_min[2])

    gmsh.model.mesh.field.setNumber(field, "XMax", contact_box_max[0])
    gmsh.model.mesh.field.setNumber(field, "YMax", contact_box_max[1])
    gmsh.model.mesh.field.setNumber(field, "ZMax", contact_box_max[2])

    # Set the box field as the background mesh field
    gmsh.model.mesh.field.setAsBackgroundMesh(field)

    # 4. generate 3D mesh
    gmsh.model.mesh.generate(3)

    # 5. export mesh to .msh file
    gmsh.option.setNumber("Mesh.MshFileVersion", msh_version)
    gmsh.write(out_msh_path)

    print(f"Saved mesh to: {out_msh_path}")

    if show_gui:
        gmsh.fltk.run()

    gmsh.finalize()
```

# Main

```python
current_dir = os.getcwd()
parent_dir = os.path.dirname(current_dir)
step_path = os.path.join(parent_dir, "mesh", 'ad_gripper_less_stiff.STEP')
save_path = os.path.join(parent_dir, "mesh", 'ad_gripper_less_stiff_refined.msh')
```

```python

generate_refined_tet_mesh(
    step_path=step_path,
    out_msh_path=save_path,

    # 远离 contact region 的 mesh size
    global_size=1.0,

    # contact region 内部 mesh size
    contact_size=0.75,

    # 这里改成你 applied force box 的范围
    contact_box_min=(18., -11.0, 3.),
    contact_box_max=(19.0,  11.0, 96.0),

    show_gui=False,
)
```
